# 1. Introduction   


## 1.1 Bacground
In 2016, Cyclistic launched a successful bike-share offering. Since then, the program has grown to a fleet of 5,824 bicycles that are geotracked and locked into a network of 692 stations across Chicago. The bikes can be unlocked from one station and returned to any other station in the system anytime.

Until now, Cyclistic’s marketing strategy relied on building general awareness and appealing to broad consumer segments. One approach that helped make these things possible was the flexibility of its pricing plans: single-ride passes, full-day passes, and annual memberships. Customers who purchase single-ride or full-day passes are referred to as casual riders. Customers who purchase annual memberships are Cyclistic members.

Cyclistic’s finance analysts have concluded that annual members are much more profitable than casual riders.

## 5.3 Station Usage Level Vs. Customer Type

+ **The number of trips by customer type and station usage level**

In this part, there are more than 800 stations, I will categorize the amount of usage in four levels:

+ **top usage**: usage numbers in Top 100
+ **high usage**: usage numbers in ranking 100 to 300
+ **moderate**:  usage numbers in ranking 300 to 500
+ **low usage**: usage nubers less than 500

```python
# Checking start station numbers
new_tripdata['start_station_id'].nunique()
```

```python
# Seperate station usage ID level
station_usage = new_tripdata['start_station_id'].value_counts()
top_usage_id = station_usage[0:100].index
high_usage_id = station_usage[100:300].index
moderate_id = station_usage[300:500].index
low_usage_id = station_usage[500:].index
```

```python
# Defining usage level function
def usageLevel(x):
    if x in top_usage_id:
        return 'top_usage'
    elif x in high_usage_id:
        return 'high_usage'
    elif x in moderate_id:
        return 'moderate'
    else:
        return 'low_usage'
```

```python
# Adding new features
new_tripdata['station_usage_level'] = new_tripdata['start_station_id'].apply(usageLevel)
```

```python
new_tripdata.head()
```

```python
# Create the new aggregation table about station and customer
station_usage_customers = new_tripdata.groupby(['station_usage_level','member_casual'])['trip_duration(min)'].agg([('trip_num', np.count_nonzero),('median_trip_duration(min)', np.median), ('trip_duration_std',np.std)])
station_usage_customers.reset_index(inplace=True)

station_list_sort = ['top_usage', 'high_usage', 'moderate', 'low_usage']
station_usage_customers['station_usage_level'] = station_usage_customers['station_usage_level'].astype('category').cat.set_categories(station_list_sort)
station_usage_customers.sort_values(by='station_usage_level', inplace=True)
station_usage_customers.reset_index(drop=True, inplace=True)

station_customers_percentage = station_usage_customers.groupby('station_usage_level')['trip_num'].transform(sum)
station_usage_customers['customers_percentage'] = round(station_usage_customers['trip_num'] / station_customers_percentage, 2)
station_usage_customers
```

```python
station_percentage_list = station_usage_customers.sort_values(by=['member_casual','station_usage_level'])['customers_percentage'].reset_index(drop=True)

fig, ax = plt.subplots(figsize=(13,8))
sns.barplot(x='station_usage_level',y='trip_num', hue='member_casual', data=station_usage_customers)

for i, p in enumerate(ax.patches):
    ax.annotate(format(round(p.get_height()/1000), '.0f')+'K'+ '(' +'{:.0%}'.format(station_percentage_list[i])+')',
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'bottom', 
                   size=15,
                   xytext = (0, 0), 
                   textcoords = 'offset points')

plt.xlabel('Station_usage_level', size=16)
plt.ylabel('Numbers', size=16)
plt.title('The number of trips by customer type and station usage level', size=18)
plt.tight_layout()
```

Describe: Except for Low usage stations, all of the stations' number of members trips are more than casuals'. especially for high usage stations, members trips are more than casuals trips 20 percentage points. For the top 100 usage stations, the number of casuals' trips is very close to the members' trips. 

Note: We don't know the numbers of each type of bike in the different stations and this distribution of the population of members and casuals in each station.  So it is hard to get a specific strategy to increase membership. But top usage stations are the best area for advertising.  A large number of casual users around these areas have the best potential to be converted to membership. 

Outout: The number of casual users trips is very close to memberships' at the top 100 usage stations. It is the best place to do the advertising.
